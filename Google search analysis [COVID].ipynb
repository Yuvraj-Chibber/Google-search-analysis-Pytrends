### Projects aims to analyse different search requests made by users related to Covid.

```python
#importing libraries required
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import pytrends
```

```python
# Creating a TrendReq object
trend=TrendReq()
```

```python
# Adding trends related to keyword pair Omicron and Covid 19 to TrendReq object.
trend.build_payload(kw_list=['Omicron','Covid-19'])
```

```python
# Analysing the frequency of search request by different region
fam=trend.interest_by_region()
fam=fam.sort_values(by='Omicron',ascending=False)
fam.head(10)
```

 ### The keyword Omicron was most searched in Taiwan , Italy and Spain

```python
fam=fam.head(10)
fam.reset_index().plot(x='geoName',y=['Omicron','Covid-19'],kind='bar',figsize=(20,15))
plt.style.use('seaborn-dark')
plt.show()
```

### Barplot of different countries vs frequency of keyword search

```python
# Building an object which stores how many times 'Omicron' was searched in India during the last 3 months.
new_var=TrendReq(hl='en-US', tz=360)
new_var.build_payload(kw_list=['Omicron'],geo='IN',timeframe='today 3-m')
```

```python
new_var.interest_over_time()
```

```python
new_var=new_var.interest_over_time()
```

```python
new_var['Omicron'].plot(figsize=(20,15))
plt.style.use('dark_background')
plt.title('Total Google Searches for Omicron in India in last 3 months', fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Count')
plt.show()
```

### The plot shows that the frequency of the keyword 'Omicron' increased during December end when the third wave was arriving in India

```python
# Building an object which stores how many times keyword pair 'Omicron' and 'third wave' was searched in India during the last 3 months.
new_info=TrendReq(hl='en-US', tz=360)
new_info.build_payload(kw_list=['Omicron','third wave'],geo='IN',timeframe='today 3-m')
```

```python
new_info=new_info.interest_over_time()
```

```python
new_info.reset_index().plot(x='date',y=['Omicron','third wave'],kind='bar',figsize=(20,15))
plt.style.use('dark_background')
plt.show()
```

### The plot shows that the frequency of the keyword 'Omicron'and 'third wave' was maximum on 29th December end when the third wave was arriving in India

```python
#Creating an object which stores frequeny of keywords 'Covid','lockdown' in different countries over the last 3 months
lockdown=TrendReq(hl='en-US', tz=360)
lockdown.build_payload(kw_list=['Covid','lockdown'],timeframe='today 3-m')
```

```python
lockdown=lockdown.interest_by_region()
lockdown=lockdown.sort_values(by='lockdown',ascending=False)
lockdown.head(10)
```

### Keyword 'lockdown'  was searched most frequently in Austria , Netherlands , Germany and Greece

```python
lockdown=lockdown.head(10)
lockdown.reset_index().plot(x='geoName',y=['Covid','lockdown'],kind='bar',figsize=(15,7))
plt.style.use('dark_background')
plt.xlabel('Top 10 countries')
plt.ylabel('Searches')
plt.show()
```
